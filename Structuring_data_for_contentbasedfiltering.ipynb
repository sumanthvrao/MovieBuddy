```python
import http.client
import ast
import json 

import numpy as np
import pandas as pd
df=pd.read_csv("ml-100k/item_with_id.csv")
```

```python
def get_director(ide):
    director=""
    actor1=""
    actor2=""
    try:
        conn = http.client.HTTPSConnection("api.themoviedb.org")
#         print(conn)

        payload = "{}"

        conn.request("GET", "/3/movie/"+str(ide)+"/credits?api_key=993f4970725ec572407558498c67a341", payload)
        
        
        res = conn.getresponse()
        print(res)
        data = res.read()
        val=(data.decode("utf-8"))
        val=json.loads(val)
        cast=val['cast']
        crew= val['crew']
        
#         print(crew)
        for i in crew:
            if i['job']=='Director':
                director=i['name']
                break
        actor1=cast[0]['name']
        actor2=cast[1]['name']
#         print(director,actor1,actor2)
        return director,actor1,actor2
            
    except:
        print(ide)
        return director,actor1,actor2
```

```python
# df['director']=df['tmdb_id'].apply(lambda x: get_director(x) )
```

```python
# .lower().strip().replace(" ", "")
df['director'],  df['actor1'], df['actor2'] = zip(*df['tmdb_id'].apply(get_director))
```

```python
df.to_csv('ml-100k/item_cc.csv', encoding='utf-8', index=False)
```
