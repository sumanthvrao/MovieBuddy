```python
def get_credits(ide):
    conn = http.client.HTTPSConnection("api.themoviedb.org")

    payload = "{}"

    conn.request("GET", "/3/movie/"+str(ide)+"/credits?api_key=993f4970725ec572407558498c67a341", payload)

    res = conn.getresponse()
    data = res.read()
    val=(data.decode("utf-8"))
    val=json.loads(val)
    return val
```

```python
def movie_id(x):
    try:
        conn = http.client.HTTPSConnection("api.themoviedb.org")
        movie=x[0]
        year=x[1]
        payload = "{}"

        conn.request("GET", "/3/search/movie?include_adult=false&page=1&language=en-US&append_to_response=credits&api_key=993f4970725ec572407558498c67a341&query="+movie+"&year="+year, payload)

        res = conn.getresponse()
        data = res.read()

        val=(data.decode("utf-8"))
        val=json.loads(val)
   
        id1=(val['results'][0]['id'])
           
        return id1
    except:

        
        return -1
```

```python
df= pd.DataFrame()
item_df.head()
df['movie_title']=item_df['movie_title'].apply(lambda x: str(x[:-6]))
df['year']=item_df['release_date'].astype(str)
df['year']=df['year'].apply(lambda x: str(x[-4:]))
df['movie+title']=df['movie_title'].apply(lambda x: "+".join(x.split()) )
df['tmdb_id']=df[['movie+title','year']].apply(movie_id,axis=1)

```
